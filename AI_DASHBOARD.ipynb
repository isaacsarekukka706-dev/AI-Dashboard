```python
import pandas as pd
import numpy as np

np.random.seed(42)

students = 100

data = {
    "Student_ID": range(1, students + 1),
    "Gender": np.random.choice(["Male", "Female"], students),
    "Class": np.random.choice(["A", "B", "C"], students),
    "Marks": np.random.randint(35, 100, students),
    "Attendance": np.random.randint(50, 100, students),
    "Assignments": np.random.randint(1, 11, students)
}

df = pd.DataFrame(data)

df.to_csv("student_data.csv", index=False)

print("Dataset Created Successfully")
df.head()
```

```python
%%writefile app.py

import streamlit as st
import pandas as pd
import plotly.express as px

# Load Data
df = pd.read_csv("student_data.csv")

st.title("🎓 Student Performance Dashboard")

# Dataset Overview
st.subheader("Dataset Overview")
st.dataframe(df)

# KPIs
st.subheader("Key Metrics")

col1, col2, col3 = st.columns(3)

col1.metric("Total Students", len(df))
col2.metric("Average Marks", round(df["Marks"].mean(), 2))
col3.metric("Average Attendance", round(df["Attendance"].mean(), 2))

# Visualization 1
st.subheader("Marks Distribution")
fig1 = px.histogram(df, x="Marks")
st.plotly_chart(fig1)

# Visualization 2
st.subheader("Attendance Distribution")
fig2 = px.histogram(df, x="Attendance")
st.plotly_chart(fig2)

# Visualization 3
st.subheader("Students by Gender")
fig3 = px.pie(df, names="Gender")
st.plotly_chart(fig3)

# Visualization 4
st.subheader("Students by Class")
fig4 = px.bar(df["Class"].value_counts())
st.plotly_chart(fig4)

# Visualization 5
st.subheader("Marks vs Attendance")
fig5 = px.scatter(df, x="Attendance", y="Marks", color="Gender")
st.plotly_chart(fig5)
```
